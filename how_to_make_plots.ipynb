# Correlation experiments

```python
%%time

# Folder with factors in it
from_folder = 'enwik-200/AM1000iter_fromGLOVE_factors_step1e-08/'

# Number of iterations on plot
MAX_ITER = 999

# Conduct experiments and save results to the dictionary
AM999_from_glove = datasets_corr(model_enwik, datasets_path, from_folder, MAX_ITER=MAX_ITER, plot_corrs=True)
```

```python
# Save obtained dictionary to file

temp_dict = AM999_from_glove.copy()
np.savez(open('temp.npz', 'wb'), 
         **{name:value for name,value in zip(temp_dict.keys(), temp_dict.values())})
```
